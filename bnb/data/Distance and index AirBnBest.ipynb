```python
import numpy as np
import pandas as pd
import openpyxl
import haversine as hs
```

```python
abnb = pd.read_excel('curated_data/airBnB_data.xlsx') #airBnB room
fd = pd.read_excel('curated_data/EtenDrinken_data.xlsx') #Food and drinks
mg = pd.read_excel('curated_data/MuseumGaleries_data.xlsx') #Museums and galeries
sh = pd.read_excel('curated_data/Winkelgebieden_data.xlsx') #Shopping areas
```

```python
room_abnb = 0
coor_abnb = abnb.iloc[room_abnb].values.tolist()[5:7]
```

```python
#Food and drinks

list_fd_coor = [[fd.iloc[i].values.tolist()[6:8],i] for i in range(len(fd))]
list_distance_abnb_fd = sorted([(round(hs.haversine(coor_abnb,i[0],unit='m')), i[1]) for i in list_fd_coor])

min_distance_fd = 0
max_distance_fd = 400
select_list_distance_abnb_fd = [i for i in list_distance_abnb_fd if i[0] <= max_distance_fd and i[0] >= min_distance_fd]

select_list_distance_abnb_mg
```

```python
#Musea and galleries
list_mg_coor = [[mg.iloc[i].values.tolist()[7:9],i] for i in range(len(mg))]
list_distance_abnb_mg = sorted([(round(hs.haversine(coor_abnb,i[0],unit='m')), i[1]) for i in list_mg_coor])

min_distance_mg = 0
max_distance_mg = 10000
select_list_distance_abnb_mg = [i for i in list_distance_abnb_mg if i[0] <= max_distance_mg and i[0] >= min_distance_mg]

select_list_distance_abnb_mg
```

```python
#shopping areas
list_sh_coor = [[sh.iloc[i].values.tolist()[1:3],i] for i in range(len(sh))]
list_distance_abnb_sh = sorted([(round(hs.haversine(coor_abnb,i[0],unit='m')), i[1]) for i in list_sh_coor])

min_distance_sh = 0
max_distance_sh = 10000
select_list_distance_abnb_sh = [i for i in list_distance_abnb_sh if i[0] <= max_distance_sh and i[0] >= min_distance_sh]

select_list_distance_abnb_sh
```
